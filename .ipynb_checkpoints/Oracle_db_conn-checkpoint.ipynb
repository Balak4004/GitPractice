```python
# AND operator &
```

```python
dup_jobid = df.groupby('job_id').size().reset_index(name='dupcount')
duprec = dup_jobid[dup_jobid['dupcount']==2]
duprec1 = df[df['job_id'].isin(duprec['job_id']) & (df['job_id'] == 'ST_MAN')]
duprec1
```

```python
# OR Operator |
```

```python
dup_jobid = df.groupby('job_id').size().reset_index(name='dupcount')
duprec = dup_jobid[dup_jobid['dupcount']==2]
duprec1 = df[df['job_id'].isin(duprec['job_id']) | (df['job_id'] == 'ST_MAN')]
duprec1
```
