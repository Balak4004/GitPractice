```python
import pandas as pd
from sqlalchemy import create_engine
import cx_Oracle
```

```python
mysql_engine = create_engine('oracle+cx_oracle://hr:hr@localhost:1521/xe')
connection = mysql_engine.connect()
```

```python
df1 = pd.read_sql("select * from employees1", connection)
```

```python
df1
```

```python
#method 1 using groupby and sum
```

```python
df2 = df1.groupby(['department_id', 'first_name'])['salary'].transform('sum')
df1['total_sal'] = df2
df1
```

```python
# using aggregator
```

```python
df4 = df1.groupby(['first_name','department_id']).agg(total_salary=('salary', 'sum')).reset_index()
df4
```

```python
df5 = df1.merge(df4, on = ['first_name'], how='inner')
df5
```

```python
df6 = df5[['first_name','department_id_x','salary', 'total_sal']]
df6
```

```python
df6 = df5[df5['department_id_x']==90]
df6
```
