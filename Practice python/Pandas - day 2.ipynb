```python
# group by
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df
```

```python
df.groupby('store_id')['store_id'].count()
```

```python
df.groupby(['store_id', 'region']).size().reset_index(name='count')
```

```python
df1 = df.groupby ('store_id')['quantity'].sum().reset_index()
```

```python
df1
```

```python
df2=df1['quantity'].max()
df3 = df1[df1['quantity']==df2]
df3
```

```python
df1.groupby('store_id').filter(lambda x:x['store_id'].max())
```

```python
df.groupby('store_id').filter(lambda x:x['store_id'].count()>3)
```
